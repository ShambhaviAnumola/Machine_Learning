```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("dataset_Facebook.csv",sep=';')
```

```python
df.head()
```

```python
df = df.drop('comment',axis=1)
```

```python
df = df.drop('like',axis=1)
```

```python
df = df.drop('share',axis=1)
```

```python
df =df.drop('Paid',axis=1)
```

```python
df = df.drop('Type',axis=1)
```

```python
df.head()
```

```python
sns.pairplot(df) 
plt.show()
```

```python
df['index'] = range(len(df))  # Add a new 'index' column
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
train,test = train_test_split(df,train_size= 0.7, test_size=0.3,random_state=100)
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
var_to_scale=['Lifetime People who have liked your Page and engaged with your post','Lifetime Post reach by people who like your Page']
```

```python
train[var_to_scale]=scaler.fit_transform(train[var_to_scale])
```

```python
train.head()
```

```python
y_train=train.pop('Total Interactions')
x_train=train
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.score(x_train,y_train)
```

```python
#test data
var_to_scale=['Lifetime People who have liked your Page and engaged with your post','Lifetime Post reach by people who like your Page']
test[var_to_scale]=scaler.fit_transform(test[var_to_scale])
y_test=test.pop('Total Interactions')
x_test=test
predictions=lm.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,predictions)
```

```python
y_test.shape
y_test_matrix=y_test.values.reshape(-1,1)
```

```python
dframe=pd.DataFrame({'actual':y_test_matrix.flatten(),'Predicted':predictions.flatten()})
```

```python
fig=plt.figure()
plt.scatter(y_test,predictions)
plt.title('Actual versus Predictions')
plt.xlabel('Actual', fontsize=20)
plt.ylabel('Predicted', fontsize=20)
```
