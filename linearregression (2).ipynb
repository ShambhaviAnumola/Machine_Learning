```python
import pandas as pd
data=pd.read_csv("HOUSE1.csv")
```

```python
data.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
data.info()
```

```python
data.describe(include='all')
```

```python
cat=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
def binary_map(x):
    return x.map({'yes':1,'no':0})
data[cat]=data[cat].apply(binary_map)
```

```python
data.head()
```

```python
table=pd.get_dummies(data['furnishingstatus'])
table.head()
```

```python
table=pd.get_dummies(data['furnishingstatus'],drop_first=True)
table.head()
```

```python
data=pd.concat([data,table],axis=1)
data.head()
```

```python
data.drop(['furnishingstatus'],axis=1,inplace=True)
data.head()
```

```python
sns.pairplot(data)
plt.show()
```

```python
data.columns
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train,df_test=train_test_split(data,train_size=0.7,test_size=0.3,random_state=100)
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
var_to_scale=['area','bedrooms','bathrooms','stories','parking','price']
```

```python
df_train[var_to_scale]=scaler.fit_transform(df_train[var_to_scale])
```

```python
df_train.head()
```

```python
y_train=df_train.pop('price')
x_train=df_train
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.score(x_train,y_train)
```

```python
#test data
var_to_scale=['area','bedrooms','bathrooms','stories','parking','price']
df_test[var_to_scale]=scaler.fit_transform(df_test[var_to_scale])
y_test=df_test.pop('price')
x_test=df_test
predictions=lm.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,predictions)
```

```python
y_test.shape
y_test_matrix=y_test.values.reshape(-1,1)
```

```python
dframe=pd.DataFrame({'actual':y_test_matrix.flatten(),'Predicted':predictions.flatten()})
```

```python
fig=plt.figure()
plt.scatter(y_test,predictions)
plt.title('Actual versus Predictions')
plt.xlabel('Actual', fontsize=20)
plt.ylabel('Predicted', fontsize=20)
```

```python
sns.regplot(y_test,predictions)
plt.title('Actual versus Predictions')
plt.xlabel('Actual', fontsize=20)
plt.ylabel('Predicted', fontsize=20)
```
